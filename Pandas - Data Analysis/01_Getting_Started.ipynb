```python
#Describe - Summary of the csv file
nba_csv.describe()
```

```python
#Frequency or Number of Occurrences of each city
nba_csv.City.value_counts()
```

```python
#Those values in the CSV File where the value of the EXP Column is 4
#Analogy with SQL - SELECT * FROM nba_csv Table where EXP value == 4
nba_csv[nba_csv.EXP == 4]
```

```python
#Those values in the CSV File where the value of the EXP Column is 4 & the POS column is 'F'
#Analogy with SQL - SELECT * FROM nba_csv Table where EXP value == 4 & POS == 'F'
nba_csv[(nba_csv["POS"] == "F") & (nba_csv.EXP == 4)]
```

```python
# Calculate a new column from an existing column
nba_csv["Ht (Cm.)"] = nba_csv["Ht (In.)"] * 2.54
nba_csv.head()
```
